## Second Method

```python
data = {
    "Player": np.array([
        "B Azam",
        "M Rizwan",
        "F Zaman",
        "Imam-ul-Haq",
        "S Masood"
    ]),
    "2020_Runs": np.array([984, 868, 550, 624, 421]),
    "2021_Runs": np.array([1024, 1326, 730, 912, 687]),
    "2022_Runs": np.array([1184, 996, 1016, 1032, 890]),
    "2023_Runs": np.array([976, 845, 642, 788, 754])
}
data1=pd.DataFrame(data=data)
data1
```

```python
plt.bar(data1['Player'],data1['2020_Runs'],width=0.5)
```

```python
plt.bar(np.arange(data1.shape[0]) - 0.2,data1['2020_Runs'],width=0.2,color='red')
plt.bar(np.arange(data1.shape[0]),data1['2021_Runs'],width=0.2)
plt.bar(np.arange(data1.shape[0]) + 0.2 ,data1['2020_Runs'],width=0.2)
# there are some Technical thing Init figure it
# add arrange and select first column by using shape[0]
# add move forward and backwart add the some value or substract some value
# 1 --------> also adjust the with parameter



plt.xticks(np.arange(data1.shape[0]),data1['Player']) 
# add the batman name at x-axis where number show so using function
# there are two parameter are passed   --1 --- existance lable and ---2----now nae which show on graph



# name underthe chart are Overlaps and Zero Readabilty 
# make its rotation is verticle  labay nazr ata hein
plt.xticks(rotation='vertical')


```

```python
# add bottom parameter add the value over the other value which are before Presented 
# bottom ---- ad upper value column than goe on
plt.bar(data1['Player'],data1['2020_Runs'],label='2020')
plt.bar(data1['Player'],data1['2021_Runs'], bottom=data1['2020_Runs'],label='2021')
plt.bar(data1['Player'],data1['2022_Runs'],bottom=data1['2020_Runs'] + data1['2021_Runs'],label='2022')
```

# SEABORN:
     Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive, informative, and statistical graphics with very little code.

### In simple terms:

👉 Matplotlib = very powerful but more manual work
👉 Seaborn = easier, cleaner, designed for data analysis

✅ Better-looking graphs by default

✅ Requires less code

✅ Designed for data analysis & EDA

✅ Works directly with Pandas DataFrames

✅ Great for distributions, relationships, correlations

In short: Seaborn = easier + cleaner + more informative plots.

# 1 Scatter Plot :

```python
# reload full penguins dataset to ensure numeric columns are present
penguain = sns.load_dataset('penguins')

# drop rows with missing values in the two columns we want to plot
penguain = penguain.dropna(subset=['flipper_length_mm', 'body_mass_g'])

# scatter Plot between flipper_length_mm and body_mass_g
sns.scatterplot(data=penguain, x='flipper_length_mm', y='body_mass_g')
```

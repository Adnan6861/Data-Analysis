
### What is Data Normalization?

**Data normalization means making numbers comparable** by putting them on a **similar scale**.

👉 Nothing fancy. Just **balancing numbers**.

---

### Real-life example (VERY important)

Imagine you’re comparing two students:

| Student | Height (cm) | Weight (kg) |
| ------- | ----------- | ----------- |
| A       | 170         | 65          |
| B       | 180         | 70          |

Now imagine a computer:

* Height values ≈ **170–180**
* Weight values ≈ **60–70**

Height numbers are **bigger**, so the computer thinks:

> “Height is more important than weight”

❌ **That’s wrong thinking**

---

### Why normalization is needed

Without normalization:

* Bigger numbers **dominate**
* Smaller numbers get **ignored**
* Model gives **wrong results**

Normalization fixes this.

---

### Type 1️⃣ Scaling (Min-Max Normalization)

#### Idea:

Bring values between **0 and 1**

#### Formula (don’t panic 😄):

```
(value − min) / (max − min)
```

#### Example:

Length values:

```
150, 160, 170
```

After scaling:

```
0.0, 0.5, 1.0
```

Now all values are **balanced**.

---

### Type 2️⃣ Centering (Standardization)

#### Idea:

Make:

* Average = **0**
* Spread = **1**

#### Formula:

```
(value − mean) / standard deviation
```

#### Example:

If mean = 170

| Original | After centering |
| -------- | --------------- |
| 160      | -1              |
| 170      | 0               |
| 180      | +1              |

Now:

* Negative = below average
* Positive = above average


### When do we use normalization?

✅ When:

* Features have **different units**
* Using ML models like:

  * KNN
  * Linear Regression
  * SVM
  * Neural Networks

❌ Not needed for:

* Decision Trees
* Random Forest







# Dealing with Missing Value in Python

Usually, missing value in dataset appears as “?”, “N/A”, 0 or just a blank cell.
these are the typical options you can consider:

• The first is to check if the person or group that collected the data can go back
and find what the actual value should be.

Another possibility is just to remove the data where that missing value is found.
• When you drop data, you can either drop the whole variable or just the single data
entry with the missing value.

If you don’t have a lot of observations with missing data, usually dropping the particular
entry is the best.

If you’re removing data, you want to look to do something that has the least amount
of impact.

Replacing data is better, since no data is wasted.
However, it is less accurate since we need to replace missing data with a guess of what
the data should be.

One standard replacement technique is to replace missing values by the average value of the

| Case                      | What to do        |
| ------------------------- | ----------------- |
| Numeric data missing      | Use mean / median |
| Categorical data missing  | Use mode          |
| Extra knowledge available | Fill logically    |

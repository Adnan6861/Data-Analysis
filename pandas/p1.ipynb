# how to add Dates sequence

```python
dates=pd.date_range('20260101',periods=6)
dates
```

```python
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list("ABCD"))
df
```

| Object          | Use                    |
| --------------- | ---------------------- |
| `pd.Timestamp`  | One specific date/time |
| `pd.date_range` | Multiple dates         |
| `DatetimeIndex` | Index of dates         |

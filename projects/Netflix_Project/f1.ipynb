## Add Magic in to Cloumn

```python
condation=[
    n_file['Vote_Average'] > 8,
    (n_file['Vote_Average'] < 8) &(n_file['Vote_Average']>6),
    (n_file['Vote_Average'] < 6)&(n_file['Vote_Average'] > 4)
]
choice=['Awsome','Good','Not Bad']
n_file['Vote_Average']=np.select(condation,choice,default='Worst')
```
